```python
from pathlib import Path
import numpy as np
import plotly.graph_objects as go

npz_path = Path('/home/exx/Downloads/embedding_goal.npz') ## CHANGE THE PATH
point_data = np.load(npz_path, allow_pickle = True)
sim = point_data['points_colors']
# sim = point_data
print(sim.shape) # sim (num_points,6)
sim_points = sim[:,:3]
sim_colors = sim[:,3:]

skip = 1
go_data=[
    go.Scatter3d(x=sim_points[::skip,0], y=sim_points[::skip,1], z=sim_points[::skip,2], 
                 mode='markers', marker=dict(size=1, color=sim_colors,
                 symbol='circle')),
]
layout = go.Layout(
    scene=dict(
        aspectmode='data'
    )
)

fig = go.Figure(data=go_data, layout=layout)
fig.update_layout(scene = dict(
                                    xaxis = dict(
                                         backgroundcolor="rgba(0, 0, 0,0)",
                                         gridcolor="white",
                                         showbackground=True,
                                         zerolinecolor="white",
                                        showticklabels = False),
                                    yaxis = dict(
                                        backgroundcolor="rgba(0, 0, 0,0)",
                                        gridcolor="white",
                                        showbackground=True,
                                        zerolinecolor="white",
                                    showticklabels = False),
                                    zaxis = dict(
                                        backgroundcolor="rgba(0, 0, 0,0)",
                                        gridcolor="white",
                                        showbackground=True,
                                        zerolinecolor="white",
                                    showticklabels = False),),
                     )
fig.show()
```
